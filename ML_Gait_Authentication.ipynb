```python
def data_collection(dataframe):
  count = 0
  x_acc = []
  y_acc = []
  z_acc = []
  extracted_features = []



  for value in range(len(dataframe['x'])):
    #print(value)
    if count == 4:
      avg_x = mean(x_acc)
      avg_y = mean(y_acc)
      avg_z = mean(z_acc)

      var_x = variance(x_acc)
      var_y = variance(y_acc)
      var_z = variance(z_acc)

      data_tuple = (avg_x, avg_y, avg_z, var_x, var_y, var_z)
      extracted_features.append(data_tuple)

      x_acc.clear()
      y_acc.clear()
      z_acc.clear()

      count = 0
    else:
      x_acc.append(dataframe['x'][value])
      y_acc.append(dataframe['y'][value])
      z_acc.append(dataframe['z'][value])
      count = count + 1


  return extracted_features
```

```python
U1 = data_collection(df1)
U2 = data_collection(df2)
U3 = data_collection(df3)
features = U1, U2, U3
```
